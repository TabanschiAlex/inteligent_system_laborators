```python
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
# a. Citiți imaginile din aceste fișiere și salvați-le într-un np.array (va avea dimensiunea 9x400x600).
images = np.array([np.load(r"images/car_{index}.npy".format(index = i)) for i in range(9)])
print(images)

united_images = np.concatenate(images)
io.imshow(united_images.astype(np.uint8))
```

```python
# b. Calculați suma valorilor pixelilor tuturor imaginilor.
pixels_sum = np.sum(images)
print(pixels_sum)
```

```python
# c. Calculați suma valorilor pixelilor pentru fiecare imagine în parte.
images_pixels_sum = np.sum(images, axis = (1, 2))
print(images_pixels_sum)
```

```python
# d. Afișați indexul imaginii cu suma maximă.
images_pixels_sum_max_position = np.argmax(images_pixels_sum)
print(images_pixels_sum_max_position)
```

```python
# e. Calculați imaginea medie și afișati-o.
mean_image = np.mean(images, axis = 0)
io.imshow(mean_image.astype(np.uint8))
io.show()
```

```python
# f. Cu ajutorul funcției np.std(images_array), calculați deviația standard a imaginilor.
image_standard_deviation  = np.std(images)
print(image_standard_deviation)
```

```python
# g. Normalizați imaginile. (se scade imaginea medie și se împarte rezultatul la deviația standard)
normalized_images = (images - mean_image) / np.std(images)
print(normalized_images)
```

```python
# h. Decupați fiecare imagine, afișând numai liniile cuprinse între 200 și 300, respectiv coloanele cuprinse între 280 și 400.
crop = images[:, 200:301, 290:401]
plt.imshow(images[7].astype(np.uint), cmap = 'gray')
plt.imshow(crop[7].astype(np.uint), cmap = 'gray')
```
